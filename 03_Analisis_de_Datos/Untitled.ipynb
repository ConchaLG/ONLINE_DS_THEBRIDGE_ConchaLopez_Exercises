```python


import sqlite3 

# Conectar a la base de datos
conn = sqlite3.connect('path_to_your_database.db')  # Cambia por el path correcto
cursor = conn.cursor()

1. Recuperar el informe de la escena del crimen 
SELECT * 
FROM crime_reports 
WHERE crime_type = 'Murder' AND crime_date = '2018-01-15';

2. Verificar los sospechosos relacionados con el asesinato
SELECT s.*
FROM suspects s
JOIN crime_reports cr ON s.crime_id = cr.id
WHERE cr.crime_type = 'Murder' AND cr.crime_date = '2018-01-15';

3. Obtener testigos del crimen
SELECT w.*
FROM witnesses w
JOIN crime_reports cr ON w.crime_id = cr.id
WHERE cr.crime_type = 'Murder' AND cr.crime_date = '2018-01-15';

4. Investigar las evidencias encontradas
SELECT e.*
FROM evidence e
JOIN crime_reports cr ON e.crime_id = cr.id
WHERE cr.crime_type = 'Murder' AND cr.crime_date = '2018-01-15';

5: Ejecutar las Consultas
# Ejecutar la consulta para el informe del crimen
cursor.execute("SELECT * FROM crime_reports WHERE crime_type = 'Murder' AND crime_date = '2018-01-15';")
crime_report = cursor.fetchall()

# Ejecutar la consulta para los sospechosos
cursor.execute("SELECT s.* FROM suspects s JOIN crime_reports cr ON s.crime_id = cr.id WHERE cr.crime_type = 'Murder' AND cr.crime_date = '2018-01-15';")
suspects = cursor.fetchall()

# Ejecutar la consulta para testigos
cursor.execute("SELECT w.* FROM witnesses w JOIN crime_reports cr ON w.crime_id = cr.id WHERE cr.crime_type = 'Murder' AND cr.crime_date = '2018-01-15';")
witnesses = cursor.fetchall()

# Ejecutar la consulta para evidencias
cursor.execute("SELECT e.* FROM evidence e JOIN crime_reports cr ON e.crime_id = cr.id WHERE cr.crime_type = 'Murder' AND cr.crime_date = '2018-01-15';")
evidence = cursor.fetchall()

6: Análisis de Resultados
Una vez que tengas los resultados, analiza la información obtenida para determinar quiénes son los sospechosos, qué testimonios pueden ser relevantes y qué evidencias existen.

7: Presentación de Resultados
print("Informe del Crimen:", crime_report)
print("Sospechosos:", suspects)
print("Testigos:", witnesses)
print("Evidencia:", evidence)

Cierre de Conexión
conn.close()
```
