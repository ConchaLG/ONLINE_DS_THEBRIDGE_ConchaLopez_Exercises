### #1.3

### #2.1 Categricas

````python
sns.countplot(data=df_titanic, x='survived')
plt.title('Supervivencia en el Titanic')
plt.show()
print(df_titanic['survived'].value_counts(normalize=True))

sns.countplot(data=df_titanic, x='who')
plt.title('Distribucin por gnero')
plt.show()
print(df_titanic['who'].value_counts(normalize=True))

sns.countplot(data=df_titanic, x='embarked')
plt.title('Puerto de embarque')
plt.show()
print(df_titanic['embarked'].value_counts(normalize=True))

sns.countplot(data=df_titanic, x='alone')
plt.title('Viajar solo o acompaado')
plt.show()
print(df_titanic['alone'].value_counts(normalize=True))
````

### #2.2 Numricas

````python
sns.histplot(df_titanic['age'], bins=30, kde=True)
plt.title('Distribucin de la Edad')
plt.show()
print(df_titanic['age'].describe())


sns.histplot(df_titanic['fare'], bins=30, kde=True)
plt.title('Distribucin de la Tarifa')
plt.show()
print(df_titanic['fare'].describe())


sns.countplot(data=df_titanic, x='pclass')
plt.title('Distribucin de clases')
plt.show()
print(df_titanic['pclass'].value_counts(normalize=True))
````
