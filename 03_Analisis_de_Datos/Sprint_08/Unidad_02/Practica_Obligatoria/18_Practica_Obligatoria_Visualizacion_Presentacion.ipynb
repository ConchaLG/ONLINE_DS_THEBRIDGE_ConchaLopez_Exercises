```python
def scatter_plot_controls(df,x_col,y_col,color_col = None, size_col= '', scale = 1, legend = True):
    fig,ax = plt.subplots(figsize = (20,10))
    sizes = (10,1000)
    if type(size_col) == str:
        if size_col != '':
            if color_col:
                scatter = sns.scatterplot(x = x_col, y = y_col, 
                                          size = df[size_col] * scale, 
                                          data = df, 
                                          hue = color_col, 
                                          palette = 'viridis', 
                                          ax = ax, 
                                          sizes = sizes, 
                                          alpha = 0.5,
                                        )
            else:
                sns.scatterplot(x = x_col, y = y_col, 
                                          size = df[size_col] * scale, 
                                          data = df, 
                                          palette = 'viridis', 
                                          ax = ax, 
                                          sizes = sizes, 
                                          alpha = 0.5,
                                        )
        elif color_col:
            scatter = sns.scatterplot(x = x_col,
                                      y = y_col,
                                      data = df,
                                      hue = color_col,
                                      alpha = 0.5,
                                      palette = 'viridis',
                                      ax = ax,
                                      sizes = sizes
                                     )
        else:
            scatter = sns.scatterplot(x = x_col, 
                                      y = y_col, 
                                      size = size_col, 
                                      data = df, 
                                      alpha = 0.5, 
                                      palette = 'viridis',
                                      ax = ax,
                                      sizes = sizes
                                     )
        if legend:
            ax.legend(ncol = 2, 
                      loc = 'upper right', 
                      frameon = False, # marco 
                      labelspacing = 2 # de dos en dos
                     )
    
        ax.set_xlabel(x_col)
        ax.set_ylabel(y_col)

        return ax




    
```

```python
df['Child'] = df['age'] < 18  
df_filtered = df[(df['sex'] == 'female') | (df['Child'])] 

survival_rate = df_filtered.groupby(['pclass', 'sex']).agg(
    Survived=('survived', 'mean'),
    Count=('survived', 'size')
).reset_index()

children_survival_rate = df_filtered[df_filtered['Child']].groupby('pclass').agg(
    Survived=('survived', 'mean'),
    Count=('survived', 'size')
).reset_index()
children_survival_rate['sex'] = 'child'

final_survival_rate = pd.concat([survival_rate, children_survival_rate], ignore_index=True)
final_survival_rate = final_survival_rate[final_survival_rate['sex'].isin(['female', 'child'])]

plt.figure(figsize=(10, 6))

for sex in final_survival_rate['sex'].unique():
    subset = final_survival_rate[final_survival_rate['sex'] == sex]
    plt.bar(subset['pclass'].astype(str) + ' - ' + sex, subset['Survived'], label=sex)

plt.title('Tasa de Supervivencia por Clase y Género (Sin hombres)')
plt.xlabel('Clase y Género')
plt.ylabel('Tasa de Supervivencia')
plt.ylim(0, 1)
plt.legend()
plt.xticks(rotation=45) 
plt.grid(axis='y')
plt.show()
```

```python
df['Child'] = df['age'] < 18  
df_filtered = df[(df['sex'] == 'female') | (df['Child'])]  

survival_rate = df_filtered.groupby(['pclass', 'sex']).agg(
    Survived=('survived', 'mean'),
    Count=('survived', 'size')
).reset_index()

children_survival_rate = df_filtered[df_filtered['Child']].groupby('pclass').agg(
    Survived=('survived', 'mean'),
    Count=('survived', 'size')
).reset_index()
children_survival_rate['sex'] = 'child'

final_survival_rate = pd.concat([survival_rate, children_survival_rate], ignore_index=True)
final_survival_rate = final_survival_rate[final_survival_rate['sex'].isin(['female', 'child'])]

final_survival_rate['pclass'] = final_survival_rate['pclass'].replace({1: 'First', 2: 'Second', 3: 'Third'})
final_survival_rate['pclass_sex'] = final_survival_rate['pclass'] + ' - ' + final_survival_rate['sex']

plt.figure(figsize=(10, 6))


colors = ['lightgray', 'white'] 
for i, sex in enumerate(final_survival_rate['sex'].unique()):
    subset = final_survival_rate[final_survival_rate['sex'] == sex]
    plt.bar(subset['pclass_sex'], subset['Survived'], label=sex, color=colors[i % len(colors)])


plt.title('', color='white')
plt.xlabel('', color='white')
plt.ylabel('', color='white')
plt.ylim(0, 1)
plt.legend(facecolor='none', fontsize='medium', frameon=False, labelspacing=1) 
plt.setp(plt.gca().get_legend().get_texts(), color='white')
plt.xticks(rotation=45, color='white') 
plt.yticks(color='white') 
plt.gca().set_facecolor('none') 
plt.gcf().patch.set_facecolor('none')
plt.grid(axis='y', color='white', linestyle='--', alpha=0.5) 
plt.xticks([]) 
plt.yticks([]) 

plt.show()
```
